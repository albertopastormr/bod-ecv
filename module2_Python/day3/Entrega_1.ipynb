La columna `Puesto` indica el tipo de puesto en un determinado centro. Podemos preguntar por los distintos tipos de puestos que existen (distintos valores de esta columna) junto con la frecuencia en la que aparece en el DataFrame. Usamos la función __value_counts__.

```python
# Sol :
# Relación de puestos de trabajo.
datos['Puesto'].value_counts()
```

## Ej

Crea un nuevo dataframe con las columnas:
Puesto, Grupo, Cat. Profesional, EUROS, Formación y Localidad. 

```python
# Sol : 
nuevos_datos = datos.loc[:,['Puesto', 'Grupo','Cat. Profesional','EUROS','Formación','Localidad\nOtras Características']]
nuevos_datos.head()
```

## Ej

Modifica el dataframe creado en el apartado anterior. Baja el sueldo en un 10% a todos los puestos de trabajo.

```python
# Sol : 
nuevos_datos['EUROS'] *= 0.9
nuevos_datos.head()
```

```python
nuevos_datos['EUROS'] = nuevos_datos['EUROS'].apply(round)
nuevos_datos.head()
```

```python
nuevos_datos['EUROS'] = nuevos_datos['EUROS'].apply(lambda x: x + 10000 if x < 10000 else x)
nuevos_datos.head()
```

## Ej

Nos interesa conocer los centros con consejeros técticos (`CONSEJERO/A TÉCNICO/A`) que no tienen titulación (valor NAN en la columna `Titulación`). Podemos anidar condiciones para la selección de filas:
* Creamos dos máscaras ( dos condiciones )
* Usa la funciín __isnull__ para detectar los valores que son NAN en la tabla
* Escribimos la conjunción de ellas

```python
# Sol: 
datos[(datos['Puesto'] == 'CONSEJERO/A TÉCNICO/A') & (datos['Titulación'].isnull())] 
```

## Ej
